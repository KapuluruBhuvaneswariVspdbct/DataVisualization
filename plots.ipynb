```python
import pandas as pd
d=pd.DataFrame({
 'quarter': ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8', 'Q9', 'Q10'],
    'fridge': [120, 140, 130, 160, 190, 200, 210, 225, 240, 250],
    'washing_machine': [80, 145, 280, 195, 110, 130, 145, 160, 170, 185],
    'pan': [160, 170, 180, 160, 190, 210, 200, 220, 210, 225]
    
})
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(12, 4))

plt.plot(d['quarter'], d['fridge'], color="blue", label="Fridge")
plt.plot(d['quarter'], d['washing_machine'], color="red", label="Washing Machine")
plt.plot(d['quarter'], d['pan'], color="green", label="Pan")

plt.title("SALES GRAPH")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
total=d[["fridge","washing_machine","pan"]].sum()
id=total.index

plt.pie(total,labels=id,autopct="%1d%%",shadow=True,explode=(0.1,0,0))
plt.title("SALES PIE CHART") 
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 4))
d.plot(kind="bar",x='quarter')
```
